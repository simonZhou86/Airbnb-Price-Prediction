### Data Preprocessing

#### Detecting missing values

```python
df.isnull().values.any()
#df.isnull()
```

```python
#df.isnull().sum()
df.isnull().sum().sum()
```

#### Interpretations:

1. pd.isnull() function returns the dataframe with each entry is a boolean value. Returned True means there is a particular entry that is missing, False means is not a missing value.

2. pd.isnull().values.any() returns whether the dataframe contains missing value. In this case, the answer is True.

3. pd.isnull().sum() returns the total number of missing values for each feature(column). pd.isnull().sum().sum() returns the total number of missing values for the whole dataframe.

#### Handling missing values:

```python
df_new = df.dropna()
df_new.head()
```

```python
df_new = df[df['beds'].notna()]
df_new
```

```python
df_new = df[df['beds'].notna() & df['bedrooms'].notna()]
df_new
```

```python
df['beds'].fillna((df['beds'].mean()), inplace=True)
df['beds'].isnull().sum()
```

#### Interpretation:

1. pd.dropna() returns the dataframe after dropping the row that contains missing values.

2. df[df['beds'].notna()] only drops the row that has the missing value for feature 'beds'.

3. df[df['beds'].notna() & df['bedrooms'].notna()] drops the row that has the missing value for both feature 'beds' AND 'bedrooms'.

4. pf.fillna() returns the dataframe after falling missing values with a specific value for some specific columns. In this case, the missing values in 'beds' column is replaced by the mean value of this column. inplace = True command means to directly modify on the current dataframe. I could see now the 'bed' column does not contain any missing values.

#### Categorical encoding

```python
print(list(df.room_type)[:10])
```

```python
encoded_room_type_column = df.room_type.astype('category').cat.codes
print(list(encoded_room_type_column)[:10])
```

```python
df.room_type = encoded_room_type_column
print(list(df.room_type)[:10])
```

#### Interpretations:

1. pd.Series.cat.codes returns series of code. In this example, it returns the categorical codes for each type. df.room_type.astype('category').cat.codes operates as first change the data type of feature 'room type' to categorical, and then apply for the categorical encoding on the entire column to get the categorical code for each type(eg. entire home/apt encodes as 0 and private room encodes as 2). Updating the dataframe by using 'df.room_type = encoded_room_type_column' command.
